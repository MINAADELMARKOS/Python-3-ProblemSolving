```python
# Count Articles
articles_counts = []
for article in data['Article']:
  article_count = len(article.split())
  articles_counts.append(article_count)
data['Article_Word_Count'] = articles_counts
data.head()
```

```python
#Count Title

Titles_counts = []

for tit in data['Title']:
  Titles_count = len(tit.split())
  Titles_counts.append(Titles_count)
data['Titles_Word_Count'] = Titles_counts
data.head()
```

```python
#lambda Function

data['Article_Count_lambda'] = data['Article'].apply(lambda n:len(n.split()))
data['Title_Count_lambda'] = data['Title'].apply(lambda n:len(n.split()))
```
